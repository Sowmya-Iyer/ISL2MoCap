```
import json 
import os
from itertools import islice
# Opening JSON file 
# f = open('/content/drive/MyDrive/ISL Openpose/Output/Tea/json/tea_000000000000_keypoints.json',) 
  
# returns JSON object as  
# a dictionary 
poses=[]
faces=[]
Hands_left=[]
Hands_right=[]
def divide_chunks(l, n):   
    # looping till length l 
    for i in range(0, len(l), n):  
        yield l[i:i + n] 
for files in os.listdir("/content/drive/MyDrive/ISL Openpose/Output/Tea/json"):
  f = open("/content/drive/MyDrive/ISL Openpose/Output/Tea/json/"+files)
  # data = json.load(f) 
  data = json.load(f)
  print(files) 
  # returns JSON object as  
  # a dictionary 
  for i in data["people"]:
    Poses_temp=list(divide_chunks(i['pose_keypoints_2d'], 3)) 
    faces_temp=list(divide_chunks(i['face_keypoints_2d'], 3))
    Hands_left_temp=list(divide_chunks(i['hand_left_keypoints_2d'], 3))
    Hands_right_temp=list(divide_chunks(i["hand_right_keypoints_2d"], 3))
  # data = json.load(f) 
  for i in data["people"]:

    print("Pose " + str(len(Poses_temp))+ str(Poses_temp))
    poses.append(Poses_temp)
    print("Face " + str(len(i['face_keypoints_2d'])) + str(faces_temp))
    faces.append(faces_temp)
    print("Hand-left " + str(len(Hands_left_temp))+ str(Hands_left_temp))
    Hands_left.append(Hands_left_temp)
    print("Hand-right "+ str(len(Hands_right_temp))+ str(Hands_right_temp))
    Hands_right.append(Hands_right_temp)
f.close()
Hands_right_temp= Hands_right.copy()
Hands_left_temp = Hands_left.copy()
poses_temp= poses.copy()
```
